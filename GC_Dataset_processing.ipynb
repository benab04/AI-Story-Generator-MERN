```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import seaborn as sns


df = vibemeter_df

# Convert Vibe_Score to numeric
df['Vibe_Score'] = pd.to_numeric(df['Vibe_Score'])

# Encode the Emotion_Zone as numeric values
desired_order = [
    "Frustrated Zone",
    "Leaning to Sad Zone",
    "Sad Zone",
    "Neutral Zone (OK)",
    "Leaning to Happy Zone",
    "Happy Zone",
    "Excited Zone"
]

emotion_mapping = {emotion: i for i, emotion in enumerate(desired_order)}
df['Emotion_Zone_Encoded'] = df['Emotion_Zone'].map(lambda x: emotion_mapping.get(x, len(desired_order)))

# encoder = LabelEncoder()
# df['Emotion_Zone_Encoded'] = encoder.fit_transform(df['Emotion_Zone'])

# Mapping for interpretation
# emotion_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
print("Emotion Zone Encoding:")
for emotion, code in emotion_mapping.items():
    print(f"{emotion}: {code}")

# Prepare data for clustering
X = df[['Vibe_Score', 'Emotion_Zone_Encoded']].values

# Apply K-means clustering with exactly 2 clusters as requested
n_clusters = 2  # Fixed to 2 clusters as requested
kmeans = KMeans(n_clusters=n_clusters, init='k-means++', max_iter=300, n_init=10, random_state=42)
df['Cluster'] = kmeans.fit_predict(X)

# Visualize the clusters
plt.figure(figsize=(10, 6))
colors = ['red', 'blue']
for i in range(n_clusters):
    plt.scatter(
        X[df['Cluster'] == i, 0],
        X[df['Cluster'] == i, 1],
        s=100,
        c=colors[i],
        label=f'Cluster {i}'
    )

plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=300,
    c='yellow',
    label='Centroids'
)

plt.title('Clusters of Employees (2 Clusters)')
plt.xlabel('Vibe Score')
plt.ylabel('Emotion Zone (Encoded)')
plt.legend()
plt.tight_layout()

# Analyze the clusters
cluster_analysis = df.groupby('Cluster').agg({
    'Vibe_Score': ['mean', 'min', 'max', 'count'],
    'Emotion_Zone': lambda x: list(x)
}).reset_index()

print("\nCluster Analysis:")
print(cluster_analysis)

# Create a more interpretable visualization with actual emotion labels
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    df['Vibe_Score'],
    df['Emotion_Zone_Encoded'],
    c=df['Cluster'],
    cmap='coolwarm',  # Red-blue colormap for 2 clusters
    s=100,
    alpha=0.8
)

# Add employee IDs as labels
for i, txt in enumerate(df['Employee_ID']):
    plt.annotate(txt, (df['Vibe_Score'].iloc[i], df['Emotion_Zone_Encoded'].iloc[i]),
                 xytext=(5, 5), textcoords='offset points')

# Add emotion zone labels to y-axis
plt.yticks(list(emotion_mapping.values()), list(emotion_mapping.keys()))

plt.colorbar(scatter, label='Cluster')
plt.title('Employee Clustering by Vibe Score and Emotion Zone (2 Clusters)')
plt.xlabel('Vibe Score')
plt.ylabel('Emotion Zone')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Print cluster memberships
print("\nEmployee Clusters:")
for cluster in range(n_clusters):
    employees = df[df['Cluster'] == cluster]['Employee_ID'].tolist()
    emotions = df[df['Cluster'] == cluster]['Emotion_Zone'].tolist()
    scores = df[df['Cluster'] == cluster]['Vibe_Score'].tolist()

    print(f"Cluster {cluster}:")
    for emp, emotion, score in zip(employees, emotions, scores):
        print(f"  {emp}: Vibe Score = {score}, Emotion = {emotion}")

# Calculate silhouette score to measure cluster quality
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(X, df['Cluster'])
print(f"\nSilhouette Score: {silhouette_avg:.3f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from datetime import datetime, timedelta

# Assume the dataframes are already loaded
# vibemeter_df, performance_df, rewards_df, leave_df, activity_tracker_df, onboarding_df

def detect_vibemeter_anomalies(df):
    """
    Detect anomalies in the vibemeter dataset
    - Identifies consistently low vibe scores
    - Detects sudden drops in mood
    """
    # Create a copy to avoid modifying the original
    vibe_df = df.copy()

    # Convert emotion zone to numeric if it's not
    emotion_map = {
        'Frustrated Zone': 1,
        'Sad Zone': 2,
        'Neutral Zone (OK)': 3,
        'Leaning to Happy Zone': 4,
        'Happy Zone': 5,
        'Excited Zone': 6
    }

    # Apply mapping if column is string type
    if vibe_df['Emotion_Zone'].dtype == 'object':
        vibe_df['Emotion_Zone_Numeric'] = vibe_df['Emotion_Zone'].map(emotion_map)
    else:
        vibe_df['Emotion_Zone_Numeric'] = vibe_df['Emotion_Zone']  # Assume it's already numeric

    # Group by employee to get stats
    employee_stats = vibe_df.groupby('Employee_ID').agg(
        avg_vibe=('Vibe_Score', 'mean'),
        min_vibe=('Vibe_Score', 'min'),
        max_vibe=('Vibe_Score', 'max'),
        vibe_range=('Vibe_Score', lambda x: max(x) - min(x)),
        response_count=('Vibe_Score', 'count'),
        latest_vibe=('Vibe_Score', lambda x: x.iloc[-1] if len(x) > 0 else None)
    ).reset_index()

    # Define anomaly criteria
    # 1. Consistently low vibe (below 3)
    # 2. Recent significant drop
    # 3. Unusual pattern detected by Isolation Forest
    # 35,200,185,69
    # Feature creation for anomaly detection
    X = employee_stats[['avg_vibe', 'min_vibe', 'max_vibe', 'vibe_range', 'response_count', 'latest_vibe']]

    # Apply Isolation Forest
    iso_forest = IsolationForest(contamination=0.1, random_state=42)
    employee_stats['anomaly_score'] = iso_forest.fit_predict(X)

    # Flag employees based on criteria
    employee_stats['should_reach_out'] = (
        (employee_stats['avg_vibe'] < 3) |
        (employee_stats['latest_vibe'] <= 2) |
        (employee_stats['anomaly_score'] == -1)
    )

    return employee_stats[['Employee_ID', 'should_reach_out']]

def detect_performance_anomalies(df):
    """
    Detect anomalies in performance data
    - Low performance ratings
    - Negative feedback trends
    - Missed promotion opportunities
    """
    perf_df = df.copy()

    # Handle categorical data
    feedback_map = {
        'Exceeds Expectations': 3,
        'Meets Expectations': 2,
        'Needs Improvement': 1
    }

    # Apply mapping if needed
    if perf_df['Manager_Feedback'].dtype == 'object':
        perf_df['Feedback_Score'] = perf_df['Manager_Feedback'].map(feedback_map)

    # Group by employee
    employee_stats = perf_df.groupby('Employee_ID').agg(
        avg_rating=('Performance_Rating', 'mean'),
        min_rating=('Performance_Rating', 'min'),
        promotion_considered=('Promotion_Consideration', lambda x: any(x)),
        feedback_score=('Feedback_Score', 'mean')
    ).reset_index()

    # Define anomaly criteria
    # 1. Consistently low ratings (below 3)
    # 2. Negative feedback
    # 3. Not considered for promotion despite good performance

    # Feature matrix for anomaly detection
    X = employee_stats[['avg_rating', 'min_rating', 'feedback_score']]
    X['promotion_considered'] = employee_stats['promotion_considered'].astype(int)

    # Apply Isolation Forest
    iso_forest = IsolationForest(contamination=0.15, random_state=42)
    employee_stats['anomaly_score'] = iso_forest.fit_predict(X)

    # Flag employees based on criteria
    employee_stats['should_reach_out'] = (
        (employee_stats['avg_rating'] < 2) |
        (employee_stats['feedback_score'] <= 1.5) |
        ((employee_stats['avg_rating'] >= 4) & (~employee_stats['promotion_considered'])) |
        (employee_stats['anomaly_score'] == -1)
    )

    print("performace anomalies",len(employee_stats[employee_stats['should_reach_out']== True]))

    return employee_stats[['Employee_ID', 'should_reach_out']]

def detect_rewards_anomalies(df):
    """
    Detect anomalies in rewards data
    - No recent rewards
    - Low reward points compared to peers
    """
    rewards_df = df.copy()

    # Convert dates if needed
    if rewards_df['Award_Date'].dtype == 'object':
        rewards_df['Award_Date'] = pd.to_datetime(rewards_df['Award_Date'])

    # Calculate days since last award
    today = pd.to_datetime('today')

    # Group by employee
    employee_stats = rewards_df.groupby('Employee_ID').agg(
        total_rewards=('Reward_Points', 'sum'),
        avg_reward=('Reward_Points', 'mean'),
        award_count=('Award_Type', 'count'),
        last_award_date=('Award_Date', 'max')
    ).reset_index()

    # Calculate days since last award
    employee_stats['days_since_award'] = (today - employee_stats['last_award_date']).dt.days

    # Feature matrix for anomaly detection
    X = employee_stats[['total_rewards', 'avg_reward', 'award_count', 'days_since_award']]

    # Apply Isolation Forest
    iso_forest = IsolationForest(contamination=0.1, random_state=42)
    employee_stats['anomaly_score'] = iso_forest.fit_predict(X)

    # Define thresholds based on data distribution
    reward_threshold = employee_stats['total_rewards'].quantile(0.10)  # Bottom 10%
    print(reward_threshold, "rewards-----------------------------------")
    time_threshold = 360  # No award in last 12 months

    # Flag employees based on criteria
    employee_stats['should_reach_out'] = (
        (employee_stats['total_rewards'] < reward_threshold) |
        (employee_stats['days_since_award'] > time_threshold) |
        (employee_stats['anomaly_score'] == -1)
    )

    print("reward anomalies",len(employee_stats[employee_stats['should_reach_out']== True]))

    return employee_stats[['Employee_ID', 'should_reach_out']]

def detect_leave_anomalies(df):
    """
    Detect anomalies in leave data
    - Excessive leave
    - Unusual leave patterns
    - No leave taken (burnout risk)
    """
    leave_df = df.copy()

    # Convert dates if needed
    if leave_df['Leave_Start_Date'].dtype == 'object':
        leave_df['Leave_Start_Date'] = pd.to_datetime(leave_df['Leave_Start_Date'])
        leave_df['Leave_End_Date'] = pd.to_datetime(leave_df['Leave_End_Date'])

    # Group by employee
    employee_stats = leave_df.groupby('Employee_ID').agg(
        total_leave_days=('Leave_Days', 'sum'),
        leave_count=('Leave_Type', 'count'),
        sick_leave_ratio=('Leave_Type', lambda x: sum(x == 'Sick Leave') / len(x) if len(x) > 0 else 0),
        last_leave_date=('Leave_End_Date', 'max')
    ).reset_index()

    # Calculate days since last leave
    today = pd.to_datetime('today')
    employee_stats['days_since_leave'] = (today - employee_stats['last_leave_date']).dt.days

    # Feature matrix for anomaly detection
    X = employee_stats[['total_leave_days', 'leave_count', 'sick_leave_ratio', 'days_since_leave']]

    # Apply Isolation Forest
    iso_forest = IsolationForest(contamination=0.15, random_state=42)
    employee_stats['anomaly_score'] = iso_forest.fit_predict(X)

    # Define thresholds
    high_leave_threshold = employee_stats['total_leave_days'].quantile(0.95)  # Top 10%
    print(high_leave_threshold , "high leave threshold-------------------------------")
    no_leave_threshold = 180  # No leave in 4 months

    # Flag employees based on criteria
    employee_stats['should_reach_out'] = (
        (employee_stats['total_leave_days'] > high_leave_threshold) |
        (employee_stats['sick_leave_ratio'] > 0.7) |  # High sick leave ratio
        (employee_stats['days_since_leave'] > no_leave_threshold) |  # No leave in a long time
        (employee_stats['anomaly_score'] == -1)
    )

    print("leave anomalies",len(employee_stats[employee_stats['should_reach_out']== True]))

    return employee_stats[['Employee_ID', 'should_reach_out']]

def detect_activity_anomalies(df):
    """
    Detect anomalies in activity tracker data
    - Consistently high or low work hours
    - Unusual communication patterns
    - Significant changes in engagement
    """
    activity_df = df.copy()

    # Convert dates if needed
    if activity_df['Date'].dtype == 'object':
        activity_df['Date'] = pd.to_datetime(activity_df['Date'])

    # Group by employee
    employee_stats = activity_df.groupby('Employee_ID').agg(
        avg_work_hours=('Work_Hours', 'mean'),
        max_work_hours=('Work_Hours', 'max'),
        min_work_hours=('Work_Hours', 'min'),
        avg_messages=('Teams_Messages_Sent', 'mean'),
        avg_emails=('Emails_Sent', 'mean'),
        avg_meetings=('Meetings_Attended', 'mean'),
        activity_days=('Date', 'count')
    ).reset_index()

    # Feature matrix for anomaly detection
    X = employee_stats[['avg_work_hours', 'max_work_hours', 'min_work_hours',
                         'avg_messages', 'avg_emails', 'avg_meetings', 'activity_days']]

    # Apply Isolation Forest
    iso_forest = IsolationForest(contamination=0.15, random_state=42)
    employee_stats['anomaly_score'] = iso_forest.fit_predict(X)

    # Define thresholds based on data distribution
    high_hours_threshold = 10  # More than 10 hours average
    low_hours_threshold = 5    # Less than 5 hours average

    # Flag employees based on criteria
    employee_stats['should_reach_out'] = (
        (employee_stats['avg_work_hours'] > high_hours_threshold) |  # Potential burnout
        (employee_stats['avg_work_hours'] < low_hours_threshold) |   # Low engagement
        (employee_stats['avg_messages'] + employee_stats['avg_emails'] <
         employee_stats[['avg_messages', 'avg_emails']].mean().sum() * 0.5) |  # Low communication
        (employee_stats['anomaly_score'] == -1)
    )

    print("activity anomalies",len(employee_stats[employee_stats['should_reach_out']== True]))


    return employee_stats[['Employee_ID', 'should_reach_out']]

def detect_onboarding_anomalies(df):
    """
    Detect anomalies in onboarding data
    - Poor onboarding feedback
    - Missing mentor or training
    - Recent joiners who might need support
    """
    onboarding_df = df.copy()

    # Convert dates if needed
    if onboarding_df['Joining_Date'].dtype == 'object':
        onboarding_df['Joining_Date'] = pd.to_datetime(onboarding_df['Joining_Date'])

    # Convert feedback to numeric
    feedback_map = {
        'Poor': 1,
        'Below Average': 2,
        'Average': 3,
        'Good': 4,
        'Excellent': 5
    }

    # Apply mapping if needed
    if onboarding_df['Onboarding_Feedback'].dtype == 'object':
        onboarding_df['Feedback_Score'] = onboarding_df['Onboarding_Feedback'].map(feedback_map)

    # Calculate days since joining
    today = pd.to_datetime('today')
    onboarding_df['days_since_joining'] = (today - onboarding_df['Joining_Date']).dt.days

    # Feature matrix
    X = pd.DataFrame()
    X['Feedback_Score'] = onboarding_df['Feedback_Score']
    X['days_since_joining'] = onboarding_df['days_since_joining']
    # X['has_mentor'] = onboarding_df['Mentor_Assigned'].astype(int)
    # X['completed_training'] = onboarding_df['Initial_Training_Completed'].astype(int)

    # Apply Isolation Forest
    iso_forest = IsolationForest(contamination=0.15, random_state=42)
    onboarding_df['anomaly_score'] = iso_forest.fit_predict(X)

    # Define criteria
    onboarding_df['should_reach_out'] = (
        (onboarding_df['Feedback_Score'] <= 0)   # Average or below feedback
        # (~onboarding_df['Mentor_Assigned'] & (onboarding_df['days_since_joining'] < 90)) |  # No mentor in first 90 days
        # (~onboarding_df['Initial_Training_Completed'] & (onboarding_df['days_since_joining'] > 30)) |  # Training not completed after 30 days
        # (onboarding_df['days_since_joining'] < 60) |  # All recent joiners (< 60 days)
        # (onboarding_df['anomaly_score'] == -1)
    )
    print("onboarding anomalies",len(onboarding_df[onboarding_df['should_reach_out']== True]))

    return onboarding_df[['Employee_ID', 'should_reach_out']]

def select_employees_for_engagement(vibemeter_df, performance_df, rewards_df, leave_df, activity_tracker_df, onboarding_df):
    """
    Main function to select employees for engagement based on dataset voting
    """
    # Get anomaly results from each dataset
    vibe_results = detect_vibemeter_anomalies(vibemeter_df)
    performance_results = detect_performance_anomalies(performance_df)
    rewards_results = detect_rewards_anomalies(rewards_df)
    leave_results = detect_leave_anomalies(leave_df)
    activity_results = detect_activity_anomalies(activity_tracker_df)
    onboarding_results = detect_onboarding_anomalies(onboarding_df)

    # Get unique employee IDs across all datasets
    all_employees = set()
    for df in [vibemeter_df, performance_df, rewards_df, leave_df, activity_tracker_df, onboarding_df]:
        all_employees.update(df['Employee_ID'].unique())

    # Create voting results dictionary
    voting_results = {emp_id: {'votes': 0, 'datasets_present': 0} for emp_id in all_employees}

    # Function to update votes
    def update_votes(results_df, voting_dict):
        for _, row in results_df.iterrows():
            emp_id = row['Employee_ID']
            if emp_id in voting_dict:
                voting_dict[emp_id]['datasets_present'] += 1
                if row['should_reach_out']:
                    voting_dict[emp_id]['votes'] += 1

    # Apply voting from each dataset
    update_votes(vibe_results, voting_results)
    update_votes(performance_results, voting_results)
    update_votes(rewards_results, voting_results)
    update_votes(leave_results, voting_results)
    update_votes(activity_results, voting_results)
    update_votes(onboarding_results, voting_results)

    # Create results dataframe
    results_df = pd.DataFrame([
        {
            'Employee_ID': emp_id,
            'Datasets_Present': data['datasets_present'],
            'Positive_Votes': data['votes'],
            'Vote_Ratio': data['votes'] / data['datasets_present'] if data['datasets_present'] > 0 else 0
        }
        for emp_id, data in voting_results.items()
    ])

    # Apply selection criteria - reach out if at least half of datasets indicate need
    results_df['Should_Reach_Out'] = results_df['Vote_Ratio'] >= 0.5

    # Prioritize employees based on vote ratio and datasets present
    results_df['Priority_Score'] = results_df['Vote_Ratio'] * (1 + (results_df['Datasets_Present'] / 6))
    results_df = results_df.sort_values('Priority_Score', ascending=False)

    return results_df

def get_employee_data_availability(employee_id, all_dataframes):
    """Check which datasets contain the specified employee"""
    availability = {}

    for name, df in all_dataframes.items():
        availability[name] = employee_id in df['Employee_ID'].values

    return availability

def generate_personalized_questions(employee_id, all_dataframes, analysis_results):
    """Generate personalized questions based on available data and analysis results"""
    availability = get_employee_data_availability(employee_id, all_dataframes)
    questions = []

    # Get the specific issues detected for this employee
    employee_row = analysis_results[analysis_results['Employee_ID'] == employee_id]
    if employee_row.empty:
        return ["How are you feeling about your work experience lately?"]

    # Always start with a general question based on vibemeter if available
    if availability['vibemeter']:
        vibemeter_data = all_dataframes['vibemeter'][all_dataframes['vibemeter']['Employee_ID'] == employee_id]
        latest_mood = vibemeter_data.iloc[-1]['Emotion_Zone']

        if 'Sad' in latest_mood or 'Frustrated' in latest_mood:
            questions.append("I noticed your mood has been on the lower side recently. What's been challenging for you at work?")
        elif 'Neutral' in latest_mood or 'OK' in latest_mood:
            questions.append("I see your mood has been neutral lately. What would help improve your work experience?")
        else:
            questions.append("Your mood seems positive lately. What's been working well for you?")

    # Add specific questions based on available data

    # Performance questions
    if availability['performance']:
        performance_data = all_dataframes['performance'][all_dataframes['performance']['Employee_ID'] == employee_id]
        rating = performance_data.iloc[-1]['Performance_Rating']
        feedback = performance_data.iloc[-1]['Manager_Feedback']

        if rating < 3 or feedback == 'Needs Improvement':
            questions.append("I see there might be some challenges with your recent performance review. What support would help you perform at your best?")

        if performance_data.iloc[-1]['Promotion_Consideration'] == False and rating >= 3:
            questions.append("Despite your solid performance, I notice you weren't considered for promotion in the last cycle. How do you feel about your career progression here?")

    # Leave questions
    if availability['leave']:
        leave_data = all_dataframes['leave'][all_dataframes['leave']['Employee_ID'] == employee_id]
        total_leaves = leave_data['Leave_Days'].sum()
        last_leave = leave_data['Leave_End_Date'].max()
        today = pd.to_datetime('today')
        days_since_leave = (today - pd.to_datetime(last_leave)).days

        if days_since_leave > 90:
            questions.append(f"I noticed it's been over {days_since_leave} days since your last leave. Do you feel you need some time off to recharge?")

        if 'Sick Leave' in leave_data['Leave_Type'].values and leave_data[leave_data['Leave_Type'] == 'Sick Leave']['Leave_Days'].sum() > 10:
            questions.append("I see you've taken several sick leaves recently. Is everything okay with your health?")

    # Activity questions
    if availability['activity']:
        activity_data = all_dataframes['activity'][all_dataframes['activity']['Employee_ID'] == employee_id]
        avg_hours = activity_data['Work_Hours'].mean()

        if avg_hours > 9:
            questions.append(f"Your average work hours seem to be around {avg_hours:.1f} hours per day. Are you feeling overwhelmed with your current workload?")
        elif avg_hours < 6:
            questions.append("Your activity levels seem lower than average. Are there any barriers preventing you from fully engaging with your work?")

    # Rewards questions
    if availability['rewards']:
        rewards_data = all_dataframes['rewards'][all_dataframes['rewards']['Employee_ID'] == employee_id]
        last_award_date = rewards_data['Award_Date'].max()
        today = pd.to_datetime('today')
        months_since_award = ((today - pd.to_datetime(last_award_date)).days) / 30

        if months_since_award > 6:
            questions.append(f"It's been around {int(months_since_award)} months since your last recognition. Do you feel your contributions are being adequately recognized?")

    # Onboarding questions
    if availability['onboarding']:
        onboarding_data = all_dataframes['onboarding'][all_dataframes['onboarding']['Employee_ID'] == employee_id]
        joining_date = onboarding_data['Joining_Date'].iloc[0]
        today = pd.to_datetime('today')
        months_since_joining = ((today - pd.to_datetime(joining_date)).days) / 30

        # if months_since_joining < 3:
        #     questions.append("As you're relatively new to the team, how has your experience been so far? Is there anything that would help you feel more integrated?")

        # if not onboarding_data['Mentor_Assigned'].iloc[0]:
        #     questions.append("I notice you weren't assigned a mentor when you joined. Would having a mentor be helpful for you?")

    # Always add general well-being questions
    questions.append("What would help improve your work experience right now?")
    questions.append("Is there anything specific you'd like your manager or the team to know?")

    # Limit to 5 most relevant questions
    return questions[:5]


import shap
import matplotlib.pyplot as plt
import numpy as np
from sklearn.ensemble import RandomForestRegressor

def explain_employee_state(employee_id, all_dataframes, analysis_results):
    """
    Generate SHAP explanations for why an employee was flagged for engagement
    or what factors contribute to their current state
    """
    # Collect all available features for this employee
    employee_features = {}
    availability = get_employee_data_availability(employee_id, all_dataframes)

    # Extract features from each available dataset
    if availability.get('vibemeter', False):
        vibe_data = all_dataframes['vibemeter'][all_dataframes['vibemeter']['Employee_ID'] == employee_id]
        if not vibe_data.empty:
            employee_features.update({
                'avg_vibe_score': vibe_data['Vibe_Score'].mean(),
                'latest_vibe_score': vibe_data['Vibe_Score'].iloc[-1],
                'vibe_trend': vibe_data['Vibe_Score'].iloc[-1] - vibe_data['Vibe_Score'].iloc[0]
                    if len(vibe_data) > 1 else 0
            })

    if availability.get('performance', False):
        perf_data = all_dataframes['performance'][all_dataframes['performance']['Employee_ID'] == employee_id]
        if not perf_data.empty:
            # Create a copy to avoid SettingWithCopyWarning
            perf_data_copy = perf_data.copy()
            feedback_map = {'Exceeds Expectations': 3, 'Meets Expectations': 2, 'Needs Improvement': 1}

            if 'Manager_Feedback' in perf_data_copy.columns and perf_data_copy['Manager_Feedback'].dtype == 'object':
                perf_data_copy['Feedback_Score'] = perf_data_copy['Manager_Feedback'].map(feedback_map)

            employee_features.update({
                'avg_performance': perf_data_copy['Performance_Rating'].mean(),
                'latest_performance': perf_data_copy['Performance_Rating'].iloc[-1],
                'feedback_score': perf_data_copy['Feedback_Score'].mean() if 'Feedback_Score' in perf_data_copy.columns else 0
            })

    if availability.get('rewards', False):
        rewards_data = all_dataframes['rewards'][all_dataframes['rewards']['Employee_ID'] == employee_id]
        if not rewards_data.empty:
            today = pd.to_datetime('today')
            last_award = rewards_data['Award_Date'].max()
            days_since_award = (today - pd.to_datetime(last_award)).days

            employee_features.update({
                'total_rewards': rewards_data['Reward_Points'].sum(),
                'award_count': len(rewards_data),
                'days_since_award': days_since_award
            })

    if availability.get('leave', False):
        leave_data = all_dataframes['leave'][all_dataframes['leave']['Employee_ID'] == employee_id]
        if not leave_data.empty:
            sick_leaves = leave_data[leave_data['Leave_Type'] == 'Sick Leave']['Leave_Days'].sum() if 'Sick Leave' in leave_data['Leave_Type'].values else 0
            today = pd.to_datetime('today')
            last_leave = leave_data['Leave_End_Date'].max()
            days_since_leave = (today - pd.to_datetime(last_leave)).days

            employee_features.update({
                'total_leave_days': leave_data['Leave_Days'].sum(),
                'sick_leave_days': sick_leaves,
                'days_since_leave': days_since_leave
            })

    if availability.get('activity', False):
        activity_data = all_dataframes['activity'][all_dataframes['activity']['Employee_ID'] == employee_id]
        if not activity_data.empty:
            employee_features.update({
                'avg_work_hours': activity_data['Work_Hours'].mean(),
                'avg_messages': activity_data['Teams_Messages_Sent'].mean(),
                'avg_emails': activity_data['Emails_Sent'].mean(),
                'avg_meetings': activity_data['Meetings_Attended'].mean()
            })

    if availability.get('onboarding', False):
        onboarding_data = all_dataframes['onboarding'][all_dataframes['onboarding']['Employee_ID'] == employee_id]
        if not onboarding_data.empty:
            # Create a copy to avoid SettingWithCopyWarning
            onboarding_data_copy = onboarding_data.copy()
            joining_date = onboarding_data_copy['Joining_Date'].iloc[0]
            today = pd.to_datetime('today')
            months_since_joining = ((today - pd.to_datetime(joining_date)).days) / 30

            feedback_map = {'Poor': 1, 'Below Average': 2, 'Average': 3, 'Good': 4, 'Excellent': 5}
            if 'Onboarding_Feedback' in onboarding_data_copy.columns and onboarding_data_copy['Onboarding_Feedback'].dtype == 'object':
                onboarding_data_copy['Feedback_Score'] = onboarding_data_copy['Onboarding_Feedback'].map(feedback_map)

            employee_features.update({
                'months_since_joining': months_since_joining,
                'onboarding_feedback': onboarding_data_copy['Feedback_Score'].iloc[0] if 'Feedback_Score' in onboarding_data_copy.columns else 3
            })

    # If no features were collected, return a basic explanation
    if not employee_features:
        return {
            "important_features": [],
            "explanation_plot": None,
            "positive_factors": [],
            "negative_factors": [],
            "error": "Insufficient data for this employee"
        }

    # Handle missing values
    features_df = pd.DataFrame([employee_features])

    # Collect data for all employees to train the model
    all_employee_features = []
    all_scores = []

    for emp_id in analysis_results['Employee_ID'].unique():
        if pd.isna(emp_id):
            continue

        emp_row = analysis_results[analysis_results['Employee_ID'] == emp_id]
        if emp_row.empty:
            continue

        emp_features = {}
        emp_availability = get_employee_data_availability(emp_id, all_dataframes)

        # Extract the same features for all employees (similar to above but simplified)
        if emp_availability.get('vibemeter', False):
            vibe_data = all_dataframes['vibemeter'][all_dataframes['vibemeter']['Employee_ID'] == emp_id]
            if not vibe_data.empty:
                emp_features['avg_vibe_score'] = vibe_data['Vibe_Score'].mean()
                emp_features['latest_vibe_score'] = vibe_data['Vibe_Score'].iloc[-1]
                if len(vibe_data) > 1:
                    emp_features['vibe_trend'] = vibe_data['Vibe_Score'].iloc[-1] - vibe_data['Vibe_Score'].iloc[0]
                else:
                    emp_features['vibe_trend'] = 0

        # Similar extraction for other datasets (abbreviated for clarity)
        # In a real implementation, refactor this to avoid code duplication

        if emp_features:  # Only add if we have data
            all_employee_features.append(emp_features)
            # Get priority score from analysis results
            score = emp_row['Priority_Score'].iloc[0]
            all_scores.append(score)

    if not all_employee_features:
        return {
            "important_features": [],
            "explanation_plot": None,
            "positive_factors": [],
            "negative_factors": [],
            "error": "Insufficient data across employees for comparison"
        }

    # Create a DataFrame with all employee features
    all_features_df = pd.DataFrame(all_employee_features)

    # Ensure feature sets match
    common_features = list(set(all_features_df.columns) & set(features_df.columns))

    if not common_features:
        return {
            "important_features": [],
            "explanation_plot": None,
            "positive_factors": [],
            "negative_factors": [],
            "error": "No common features found for comparison"
        }

    # Fill NA values to avoid issues
    all_features_df = all_features_df[common_features].fillna(all_features_df[common_features].mean())
    features_df = features_df[common_features].fillna(all_features_df[common_features].mean())

    # Train a simple model (RandomForest works well with SHAP)
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    try:
        model.fit(all_features_df[common_features], all_scores)
    except Exception as e:
        return {
            "important_features": [],
            "explanation_plot": None,
            "positive_factors": [],
            "negative_factors": [],
            "error": f"Error training model: {str(e)}"
        }

    # Create the SHAP explainer
    try:
        explainer = shap.TreeExplainer(model)

        # Calculate SHAP values for the employee of interest
        shap_values = explainer.shap_values(features_df[common_features])

        # Sort features by importance
        feature_importance = pd.DataFrame({
            'Feature': common_features,
            'SHAP_Value': np.abs(shap_values).mean(0)
        }).sort_values('SHAP_Value', ascending=False)

        # Create a visual explanation (SHAP force plot)
        plt.figure(figsize=(10, 6))
        shap.force_plot(
            explainer.expected_value,
            shap_values[0],
            features_df[common_features],
            feature_names=common_features,
            matplotlib=True,
            show=False
        )

        # Save the plot
        plot_filename = f'employee_{employee_id}_shap_explanation.png'
        plt.savefig(plot_filename, bbox_inches='tight')
        plt.close()

        # Identify top contributing factors (positive and negative)
        positive_factors = []
        negative_factors = []

        for feature, shap_value in zip(common_features, shap_values[0]):
            factor = {
                'feature': feature,
                'value': float(features_df[feature].iloc[0]),
                'impact': float(abs(shap_value)),
                'direction': 'positive' if shap_value > 0 else 'negative'
            }

            if shap_value > 0:
                positive_factors.append(factor)
            else:
                negative_factors.append(factor)

        # Sort factors by impact
        positive_factors = sorted(positive_factors, key=lambda x: x['impact'], reverse=True)
        negative_factors = sorted(negative_factors, key=lambda x: x['impact'], reverse=True)

        return {
            'important_features': feature_importance.to_dict('records'),
            'explanation_plot': plot_filename,
            'positive_factors': positive_factors,
            'negative_factors': negative_factors
        }

    except Exception as e:
        return {
            "important_features": [],
            "explanation_plot": None,
            "positive_factors": [],
            "negative_factors": [],
            "error": f"Error generating SHAP explanation: {str(e)}"
        }

def generate_employee_insights(employee_id, all_dataframes, analysis_results):
    """Generate human-readable insights based on SHAP explanations"""
    explanation = explain_employee_state(employee_id, all_dataframes, analysis_results)

    insights = []

    # Check for errors
    if "error" in explanation and explanation["error"]:
        return f"Unable to generate insights: {explanation['error']}"

    # Overall summary
    insights.append("Employee State Analysis:")

    # Add positive factors
    if explanation['positive_factors']:
        insights.append("\nPositive contributing factors:")
        for i, factor in enumerate(explanation['positive_factors'][:3], 1):
            feature_name = factor['feature'].replace('_', ' ').title()
            insights.append(f"  {i}. {feature_name}: {factor['value']:.2f}")

    # Add negative factors
    if explanation['negative_factors']:
        insights.append("\nConcerning factors:")
        for i, factor in enumerate(explanation['negative_factors'][:3], 1):
            feature_name = factor['feature'].replace('_', ' ').title()
            insights.append(f"  {i}. {feature_name}: {factor['value']:.2f}")

    # Translate technical features into human insights
    translated_insights = []

    # Combine top factors from both positive and negative
    top_factors = []
    if explanation['positive_factors']:
        top_factors.extend(explanation['positive_factors'][:2])
    if explanation['negative_factors']:
        top_factors.extend(explanation['negative_factors'][:2])

    for factor in top_factors:
        feature = factor['feature']
        value = factor['value']
        direction = factor['direction']

        # Add context-specific insights based on the feature
        if feature == 'avg_vibe_score':
            if direction == 'negative' and value < 3:
                translated_insights.append(f"Employee's consistently low vibe score ({value:.1f}/6) suggests ongoing dissatisfaction.")
            elif direction == 'positive' and value > 4:
                translated_insights.append(f"Employee's high average vibe score ({value:.1f}/6) indicates generally positive engagement.")

        elif feature == 'vibe_trend':
            if value < -1:
                translated_insights.append(f"Declining vibe score trend (down {abs(value):.1f} points) suggests a recent negative shift.")
            elif value > 1:
                translated_insights.append(f"Improving vibe score trend (up {value:.1f} points) indicates positive trajectory.")

        elif feature == 'avg_work_hours':
            if value > 9.5:
                translated_insights.append(f"Unusually high work hours ({value:.1f} hrs/day) may indicate workload stress or burnout risk.")
            elif value < 6:
                translated_insights.append(f"Lower than average work hours ({value:.1f} hrs/day) could suggest disengagement or external factors.")

        elif feature == 'days_since_award' and value > 180:
            translated_insights.append(f"No recognition in {value:.0f} days may contribute to feeling undervalued.")

        elif feature == 'days_since_leave' and value > 120:
            translated_insights.append(f"No time off in {value:.0f} days could lead to burnout.")

        elif feature == 'total_rewards' and direction == 'negative':
            translated_insights.append(f"Lower recognition ({value:.0f} total reward points) compared to peers.")

        elif feature == 'avg_performance' and value < 3:
            translated_insights.append(f"Performance rating ({value:.1f}) suggests room for improvement or potential support needs.")

        elif feature == 'sick_leave_days' and value > 7:
            translated_insights.append(f"Higher sick leave usage ({value:.0f} days) might indicate health concerns or stress-related issues.")

    if translated_insights:
        insights.append("\nKey Insights:")
        for insight in translated_insights:
            insights.append(f" {insight}")

    if explanation.get('explanation_plot'):
        insights.append(f"\nRefer to visualization in: {explanation['explanation_plot']}")

    return "\n".join(insights)


def main():
    # Dictionary of all dataframes for easy access
    all_dfs = {
        'vibemeter': vibemeter_df,
        'performance': performance_df,
        'rewards': rewards_df,
        'leave': leave_df,
        'activity': activity_tracker_df,
        'onboarding': onboarding_df
    }

    # Run the analysis to select employees for engagement
    engagement_results = select_employees_for_engagement(
        vibemeter_df, performance_df, rewards_df, leave_df, activity_tracker_df, onboarding_df
    )

    # Generate report and select top priority employees
    print(f"Total employees selected for engagement: {engagement_results['Should_Reach_Out'].sum()}")
    priority_employees = engagement_results[engagement_results['Should_Reach_Out']].head(10)

    print("\nTop 10 priority employees for engagement:")
    for _, emp in priority_employees.iterrows():
        emp_id = emp['Employee_ID']
        print(f"Employee ID: {emp_id}, Priority Score: {emp['Priority_Score']:.2f}")

        # Get personalized questions for this employee
        questions = generate_personalized_questions(emp_id, all_dfs, engagement_results)
        print("Suggested questions:")
        for i, q in enumerate(questions, 1):
            print(f"  {i}. {q}")

        # Generate SHAP-based insights
        try:
            insights = generate_employee_insights(emp_id, all_dfs, engagement_results)
            print("\nKey insights based on SHAP analysis:")
            print(insights)
        except Exception as e:
            print(f"\nError generating insights: {str(e)}")

        print()

    # Create detailed report for HR
    report = engagement_results.copy()
    report['Available_Datasets'] = report['Employee_ID'].apply(
        lambda emp_id: ', '.join([name for name, df in all_dfs.items()
                                  if emp_id in df['Employee_ID'].values])
    )

    # Save report
    report.to_csv('employee_engagement_report.csv', index=False)
    print("Detailed report saved to 'employee_engagement_report.csv'")


main()
```
